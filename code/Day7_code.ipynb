```python
def getSum(x):
    output = 0
    for i in range(1,x+1):
        output += i
    return output
vfunc = np.vectorize(getSum)
```

```python
import statistics as st
import numpy as np
with open('day7_input.txt') as f:
    crab_pos = np.array([int(x) for x in f.readline().split(',')])
minFuel = 999999999999999999
for i in range(min(crab_pos), max(crab_pos)+1):
    tmp = abs(crab_pos - i)
    fuel = sum(vfunc(tmp))
    if(fuel < minFuel):
        minFuel = fuel
print(minFuel)
```
