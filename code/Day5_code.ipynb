```python
with open("day5_input_part2.txt") as f:
    lines = f.read().splitlines()
print(lines[0])
```

```python
vent_map = dict()
for x in lines:
    start, end = x.split(' -> ')
    start = [int(x) for x in start.split(',')]
    end = [int(x) for x in end.split(',')]

    if(start[0] == end[0]):
        for i in range(min(start[1], end[1]), max(start[1], end[1])+1):
            coord = str(start[0]) + ',' + str(i)
            if(coord in vent_map):
                vent_map[coord] += 1
            else:
                vent_map[coord] = 1
    elif(start[1] == end[1]):
        for i in range(min(start[0], end[0]), max(start[0], end[0])+1):
            coord = str(i) + ',' + str(start[1])
            if(coord in vent_map):
                vent_map[coord] += 1
            else:
                vent_map[coord] = 1
    else:
        for i in range(abs(end[0] - start[0])+1):
            if(start[0] < end[0]):
                xcoord = str(start[0] + i)
            else:
                xcoord = str(start[0] - i)
            if(start[1] < end[1]):
                ycoord = str(start[1] + i)
            else:
                ycoord = str(start[1] - i)
            coord = xcoord + ',' + ycoord
            if(coord in vent_map):
                vent_map[coord] += 1
            else:
                vent_map[coord] = 1
```

```python
counter = 0
for x in vent_map.values():
    if(x > 1):
        counter += 1
print(counter)
```
